```python
import pickle, json
import matplotlib.pyplot as plt

# For using core package that located in the two upper folder.
import os,sys
sys.path.append('../../')

from core.visualizer import visualize_history, plot_complexty_graph
```

```python
# Constants
SESSION_FOLDER = 'generated_files/session_1545166502.218424'
COMPLEXITIES = [10, 50]
```

```python
# Read necessary objects from disk
histories = []
infos = []
for complexity in COMPLEXITIES:
    histories.append(pickle.load(open(SESSION_FOLDER + '/histories/history_{}.save'.format(complexity), "rb")))
    infos.append(json.load(open(SESSION_FOLDER + '/models/info_{}.json'.format(complexity),)))
```

```python
def visualize_history_at(index):
    info = infos[index]
    print('{} (Epoch: {}, Batch: {}, Time Taken: {})'
          .format(info['info'], info['epoch'], info['batch'], info['time_taken']))
    visualize_history(histories[index])
```

```python
plot_complexty_graph(histories, 3)
```
